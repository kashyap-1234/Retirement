```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filtered_dataset = pd.read_csv(r'D:\A_Kashyap\UIC\Sem-1\Data_Science\Project\count_by_all_dimensions_for_high_level_model_v2.csv')
```

```python
filtered_dataset.head()
```

```python
filtered_dataset_sorted = filtered_dataset.sort_values(by='Poor Health to Dead')
plt.figure(figsize=(10, 6))
colors = plt.cm.viridis_r(range(len(filtered_dataset_sorted)))
colors = ['skyblue', 'salmon', 'lightgreen', 'lightcoral', 'gold']

plt.bar(filtered_dataset_sorted['Age'], filtered_dataset_sorted['Poor Health to Dead'], color=colors)
plt.title('Age Groups with Respect to Poor Health to Dead')
plt.xlabel('Ages')
plt.ylabel('No. of people from Poor Health to Dead')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
filtered_dataset_sorted = filtered_dataset.sort_values(by='Good Health to Dead')
scatter_color = 'blue'
plt.figure(figsize=(10, 6))
plt.scatter(filtered_dataset_sorted['Age'], filtered_dataset_sorted['Good Health to Dead'], color=scatter_color, label='Good Health to Dead', marker='o')
plt.title('Scatter Plot of Age Groups with Respect to Good Health to Dead')
plt.xlabel('Ages')
plt.ylabel('No. of people from Good Health to Dead')
plt.xticks(rotation=45, ha='right')
plt.legend()

plt.show()
```

```python
average_age = filtered_dataset.loc[filtered_dataset['Poor Health to Dead'] == 1, 'Age'].mean()

print(f'The average age of individuals who went from Poor Health to Dead is: {average_age:.2f} years')
```
