```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Loading the dataset

new_df = pd.read_csv('ds_project_complete_data.csv')
new_df
```

```python
# Cleaning the gender column

new_df['ragender'] = new_df['ragender'].apply(lambda x: x[2:])
pd.set_option('display.max_columns', None)
```

```python
ages_in_waves = new_df.loc[:, 'rabyear':'radage_y']
ages_in_waves

income_in_waves = new_df.loc[:, 'r1inc_group':'r14inc_group']

for val in income_in_waves.loc[:,'r1inc_group':'r14inc_group']:
  income_in_waves['r1inc_group'] = income_in_waves.apply(lambda row: row[val] if not pd.isnull(row[val]) else row['r1inc_group'], axis=1)
ages_in_waves['r1inc_group'] = income_in_waves['r1inc_group']

ages_in_waves['study_start_age'] = new_df['r1agey_b']
i = 0
for val in ages_in_waves.loc[:,'r1agey_b':'r14agey_b']:
  ages_in_waves['rabyear'] = ages_in_waves.apply(lambda row: (1992+(2*i))-row[val] if pd.isnull(row['rabyear']) else row['rabyear'], axis=1)
  i += 1
for val in ages_in_waves.loc[:,'r1agey_b':'r14agey_b']:
  ages_in_waves['study_start_age'] = ages_in_waves.apply(lambda row: row[val] if pd.isnull(row['study_start_age']) else row['study_start_age'], axis=1)

ages_in_waves
```

```python
ages_in_waves.dropna(subset=['study_start_age','radage_y','r1inc_group'], inplace=True)
ages_in_waves['ragender'] = new_df['ragender']
```

```python
# Creating a new dataframe with columns of Age at the start of the study, Gender, Income group, Age at Death
new_df1 = ages_in_waves[['rabyear','study_start_age','ragender','r1inc_group','radage_y']]
new_df1
```

```python
new_df1.reset_index(drop=True, inplace=True)
```

```python
# One hot encoding the column of Gender
one_hot_encoded_data = pd.get_dummies(new_df1, columns = ['ragender'])

one_hot_encoded_data.drop(columns=['ragender_female'],inplace=True)
```

```python
one_hot_encoded_data.rename(inplace=True, columns={'ragender_male':'Gender'})
one_hot_encoded_data
```

```python
# Linear regression model with a train test split of 70-30

from sklearn import datasets, linear_model, metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, f1_score, classification_report

X = one_hot_encoded_data[['study_start_age','Gender','r1inc_group']]
y = one_hot_encoded_data[['radage_y']]
X_train, X_test,y_train, y_test = train_test_split(X, y,
                                       test_size=0.3,
                                       random_state=1)


kf = KFold(n_splits=5, random_state=1, shuffle=True)
acc = 0
list_training_error = []
list_testing_error = []
for (train_index, test_index) in kf.split(X):
  reg = linear_model.LinearRegression()
  reg.fit(X.iloc[train_index], y.iloc[train_index])
  y_pred = reg.predict(X.iloc[test_index])
  y_train_data_pred = reg.predict(X.iloc[train_index])
  y_test_data_pred = reg.predict(X.iloc[test_index])

  fold_training_error = mean_absolute_error(y.iloc[train_index], y_train_data_pred)
  fold_testing_error = mean_absolute_error(y.iloc[test_index], y_test_data_pred)

  list_training_error.append(fold_training_error)
  list_testing_error.append(fold_testing_error)

  acc += accuracy_score(y.iloc[test_index], y_pred.astype(int))
avg_accuracy = acc / kf.get_n_splits(X)
print(avg_accuracy)
```

```python
# Calculating the average life expectancy, for the baseline, by averaging the Age at Death column

y_pred = reg.predict(X_test)

average_life_expectancy = np.average(one_hot_encoded_data['radage_y'])
baseline = [average_life_expectancy]*y_pred.shape[0]

#Calculating the mean squared error with respect to the baseline
y_pred_train = reg.predict(X_train)
y_pred_train = np.around(y_pred_train).astype(int)

print(mean_squared_error(y_test, baseline))
```
