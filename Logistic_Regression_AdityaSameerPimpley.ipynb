```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Loading the dataset

new_df = pd.read_csv('hrs_data_needed_for_calcs_with_inc_groups.csv')
new_df
```

```python
# Cleaning the gender column

new_df['ragender'] = new_df['ragender'].apply(lambda x: x[2:])
new_df
```

```python
# Filtering out the rows where Age at the start of the study, Age at Death and Income group is null

new_df.dropna(subset=['r8agey_b','Dead_wave_8','r8inc_group'], inplace=True)
#new_df = new_df.drop(new_df[new_df['Dead_wave_1'] == 1].index)
new_df = new_df[new_df.Dead_wave_8 != 1]
new_df
```

```python
# Creating a new dataframe with columns of Income groups,Dead in next wave,Status in other waves and Gender

new_df1 = new_df[['r8agey_b','ragender','r8inc_group','Dead_wave_9']]
new_df2 = new_df.loc[:,'Status_wave_1':'Status_wave_8']
new_df3 = new_df.loc[:,'r1inc_group':'r8inc_group']
new_df1 = pd.concat([new_df1,new_df2,new_df3],axis = 1)
new_df_0 = new_df1[new_df1['Dead_wave_9'] == 0]
new_df_0 = new_df_0.iloc[:1176,:]
new_df_1 = new_df1[new_df1['Dead_wave_9'] == 1]
new_df1 = new_df_0 + new_df_1
frames = [new_df_0, new_df_1]
new_df1 = pd.concat(frames)
new_df1
```

```python
new_df1.reset_index(drop=True, inplace=True)
new_df1
```

```python
# One hot encoding the column of Gender
one_hot_encoded_data = pd.get_dummies(new_df1, columns = ['ragender'])
one_hot_encoded_data.drop(columns=['ragender_female'],inplace=True)
one_hot_encoded_data
```

```python

one_hot_encoded_data.rename(inplace=True, columns={'ragender_male':'Gender'})
one_hot_encoded_data
```

```python
# Logistic regression model with a train test split of 70-30, also substituted Nan values with a negative value
from sklearn.linear_model import LogisticRegression
from sklearn import datasets, linear_model, metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import mean_squared_error, r2_score

X = one_hot_encoded_data.drop(columns=['Dead_wave_9'])
X = X.replace({'Healthy': 1,'HHC':2,'NH':3,'HC and NH':4})
X = X.fillna(-3)
y = one_hot_encoded_data[['Dead_wave_9']]

X_train, X_test,y_train, y_test = train_test_split(X,y,
                                       test_size=0.3,
                                       random_state=1)

iters = range(10,201,10)
accuracy=[]
for i in range(10,201,10):
  reg = linear_model.LogisticRegression(max_iter=i)
  reg.fit(X_train, y_train)
  y_pred = reg.predict(X_test)
  accuracy.append(accuracy_score(y_test,y_pred))
plt.plot(iters,accuracy)
plt.xlabel('iters')
plt.ylabel('accuracy')
plt.show
```

```python
reg = linear_model.LogisticRegression(max_iter=90)
y_pred = reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
y_pred

```
