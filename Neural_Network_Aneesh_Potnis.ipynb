```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
new_df = pd.read_csv('ds_project_complete_data.csv')
new_df
```

```python
new_df['ragender'] = new_df['ragender'].apply(lambda x: x[2:])
pd.set_option('display.max_columns', None)
new_df
```

```python
ages_in_waves = new_df.loc[:, 'rabyear':'radage_y']
ages_in_waves

income_in_waves = new_df.loc[:, 'r1inc_group':'r14inc_group']

for val in income_in_waves.loc[:,'r1inc_group':'r14inc_group']:
  income_in_waves['r1inc_group'] = income_in_waves.apply(lambda row: row[val] if not pd.isnull(row[val]) else row['r1inc_group'], axis=1)
print(income_in_waves)
ages_in_waves['r1inc_group'] = income_in_waves['r1inc_group']

ages_in_waves['study_start_age'] = new_df['r1agey_b']
i = 0
for val in ages_in_waves.loc[:,'r1agey_b':'r14agey_b']:
  ages_in_waves['rabyear'] = ages_in_waves.apply(lambda row: (1992+(2*i))-row[val] if pd.isnull(row['rabyear']) else row['rabyear'], axis=1)
  i += 1
for val in ages_in_waves.loc[:,'r1agey_b':'r14agey_b']:
  ages_in_waves['study_start_age'] = ages_in_waves.apply(lambda row: row[val] if pd.isnull(row['study_start_age']) else row['study_start_age'], axis=1)

ages_in_waves
```

```python
ages_in_waves.dropna(subset=['study_start_age','radage_y','r1inc_group'], inplace=True)
ages_in_waves['ragender'] = new_df['ragender']
ages_in_waves
```

```python
new_df1 = ages_in_waves[['rabyear','study_start_age','ragender','r1inc_group','radage_y']]
new_df1.reset_index(drop=True, inplace=True)
new_df1
```

```python
one_hot_encoded_data = pd.get_dummies(new_df1, columns = ['ragender'])
one_hot_encoded_data.drop(columns=['ragender_female'],inplace=True)
one_hot_encoded_data.rename(inplace=True, columns={'ragender_male':'Gender'})
one_hot_encoded_data
```

```python
def f(row):
    if row['radage_y'] < 65:
        val = "Less than 65"
    elif row['radage_y'] <= 74:
        val = "65 to 74"
    elif row['radage_y'] <= 84:
        val = "75 to 84"
    else:
        val = "85 plus"
    return val

one_hot_encoded_data['radage_y'] = one_hot_encoded_data.apply(f, axis=1)
one_hot_encoded_data
```

```python
one_hot_encoded_data_predict = pd.get_dummies(one_hot_encoded_data, columns = ['radage_y'])
one_hot_encoded_data_predict
```

```python
from sklearn.model_selection import train_test_split
X = one_hot_encoded_data[['study_start_age','r1inc_group','Gender']]
y = one_hot_encoded_data_predict[['radage_y_65 to 74','radage_y_75 to 84','radage_y_85 plus','radage_y_Less than 65']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

# Neural Network Model

```python
from keras.models import Sequential
from keras.layers import Dense


model = Sequential()
model.add(Dense(256, activation='relu', input_dim=3))
model.add(Dense(32, activation='relu'))

model.add(Dense(4, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy','mse'])
model.summary()
```

```python
hist = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=100)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

acc = hist.history['accuracy']
val = hist.history['val_accuracy']
epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, '-', label='Training accuracy')
plt.plot(epochs, val, ':', label='Testing accuracy')
plt.title('Training and Testing Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.xticks(np.arange(1, 21, step=2))
plt.legend(loc='lower right')
plt.plot()
```

```python
mse = hist.history['mse']
val = hist.history['val_mse']
epochs = range(1, len(mse) + 1)

plt.plot(epochs, mse, '-', label='Training MSE')
plt.plot(epochs, val, ':', label='Testing MSE')
plt.title('Training and Testing MSE')
plt.xlabel('Epoch')
plt.xticks(np.arange(1, 21, step=2))
plt.ylabel('MSE')
plt.legend(loc='upper right')
plt.plot()
```
