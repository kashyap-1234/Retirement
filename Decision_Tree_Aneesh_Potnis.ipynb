```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Loading the dataset

new_df = pd.read_csv('ds_project_complete_data.csv')
new_df
```

```python
# Cleaning the gender column

new_df['ragender'] = new_df['ragender'].apply(lambda x: x[2:])
pd.set_option('display.max_columns', None)
new_df
```

```python
ages_in_waves = new_df.loc[:, 'rabyear':'radage_y']
ages_in_waves

income_in_waves = new_df.loc[:, 'r1inc_group':'r14inc_group']

for val in income_in_waves.loc[:,'r1inc_group':'r14inc_group']:
  income_in_waves['r1inc_group'] = income_in_waves.apply(lambda row: row[val] if not pd.isnull(row[val]) else row['r1inc_group'], axis=1)
print(income_in_waves)
ages_in_waves['r1inc_group'] = income_in_waves['r1inc_group']

ages_in_waves['study_start_age'] = new_df['r1agey_b']
i = 0
for val in ages_in_waves.loc[:,'r1agey_b':'r14agey_b']:
  ages_in_waves['rabyear'] = ages_in_waves.apply(lambda row: (1992+(2*i))-row[val] if pd.isnull(row['rabyear']) else row['rabyear'], axis=1)
  i += 1
for val in ages_in_waves.loc[:,'r1agey_b':'r14agey_b']:
  ages_in_waves['study_start_age'] = ages_in_waves.apply(lambda row: row[val] if pd.isnull(row['study_start_age']) else row['study_start_age'], axis=1)

ages_in_waves
```

```python
ages_in_waves.dropna(subset=['study_start_age','radage_y','r1inc_group'], inplace=True)

ages_in_waves
```

```python
ages_in_waves['ragender'] = new_df['ragender']
ages_in_waves
```

```python
# Creating a new dataframe with columns of Age at the start of the study, Gender, Income group, Age at Death
new_df1 = ages_in_waves[['rabyear','study_start_age','ragender','r1inc_group','radage_y']]
new_df1
```

```python
new_df1.reset_index(drop=True, inplace=True)
new_df1
```

```python
# One hot encoding the column of Gender
one_hot_encoded_data = pd.get_dummies(new_df1, columns = ['ragender'])
one_hot_encoded_data.drop(columns=['ragender_female'],inplace=True)
one_hot_encoded_data
```

```python
one_hot_encoded_data.rename(inplace=True, columns={'ragender_male':'Gender'})
one_hot_encoded_data
```

```python
def f(row):
    if row['radage_y'] < 65:
        val = "Less than 65"
    elif row['radage_y'] <= 74:
        val = "65 to 74"
    elif row['radage_y'] <= 84:
        val = "75 to 84"
    else:
        val = "85 plus"
    return val

one_hot_encoded_data['radage_y'] = one_hot_encoded_data.apply(f, axis=1)
one_hot_encoded_data
```

```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split

X = one_hot_encoded_data[['study_start_age','r1inc_group','Gender']]
y = one_hot_encoded_data[['radage_y']]

# dividing X, y into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

# training a DescisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
dtree_model = DecisionTreeClassifier().fit(X_train, y_train)
y_pred = dtree_model.predict(X_test)
```

```python
def f(row):
    if row['radage_y'] < 65:
        val = "Less than 65"
    elif row['radage_y'] <= 74:
        val = "65 to 74"
    elif row['radage_y'] <= 84:
        val = "75 to 84"
    else:
        val = "85 plus"
    return val
new_df2 = pd.DataFrame()
new_df2['radage_y'] = new_df1.apply(f, axis=1)

average_life_expectancy = new_df2['radage_y'].value_counts().reset_index()
baseline = pd.DataFrame([average_life_expectancy['index'].iloc[0]]*y_pred.shape[0], columns=['radage_y'])

baseline
```

```python
y_pred = dtree_model.predict(X_test)

accuracy_score(y_pred, y_test)
```

```python
accuracy_score(y_pred, baseline)
```

```python
from sklearn.tree import DecisionTreeClassifier

training_accuracy = []
testing_accuracy = []
training_error = []
testing_error = []
for i in range(1,15):
  X = one_hot_encoded_data[['study_start_age','r1inc_group','Gender']]
  y = new_df2[['radage_y']]

  X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

  dtree_model = DecisionTreeClassifier(max_depth=i).fit(X_train, y_train)
  y_pred = dtree_model.predict(X_test)
  y_pred_train = dtree_model.predict(X_train)
  training_accuracy.append(accuracy_score(y_pred_train, y_train))
  testing_accuracy.append(accuracy_score(y_pred, y_test))

  training_error.append(1-accuracy_score(y_pred_train, y_train))
  testing_error.append(1-accuracy_score(y_pred, y_test))
```

```python
max_depth_values = []
for i in np.arange(1,15):
  max_depth_values.append(i)
plt.plot(max_depth_values, training_accuracy, label='Training Accuracy')
plt.plot(max_depth_values, testing_accuracy, label='Testing Accuracy')
plt.xticks(np.arange(1,15))
plt.xlabel("Decision tree Max Depth")
plt.ylabel("Accuracy")
plt.legend()
```

```python
max_depth_values = []
for i in np.arange(1,15):
  max_depth_values.append(i)
plt.plot(max_depth_values, training_error, label='Training Error')
plt.plot(max_depth_values, testing_error, label='Testing Error')
plt.xticks(np.arange(1,15))
plt.xlabel("Decision tree Max Depth")
plt.ylabel("Error")
plt.legend()
```

```python
dtree_model = DecisionTreeClassifier(max_depth=np.argmax(testing_accuracy), criterion='entropy').fit(X_train, y_train)
y_pred = dtree_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
